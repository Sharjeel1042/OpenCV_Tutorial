```python
import cv2
import sys
```

```python
s=0
if len(sys.argv)>1:
    s=sys.argv[1]

source=cv2.VideoCapture(s)

win_name="Camera Preview"
cv2.namedWindow(win_name,cv2.WINDOW_NORMAL)

while cv2.waitKey(1)!=27:
    has_frame,frame=source.read()
    if has_frame:
        cv2.imshow(win_name,frame)
    else:
        break

source.release()
cv2.destroyWindow(win_name)
```
