```python
import os
import cv2
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

from zipfile import ZipFile
from urllib.request import urlretrieve

matplotlib.rcParams['figure.figsize'] = (9.0, 9.0)

%matplotlib inline

def download_and_unzip(url, save_path):
    print(f"Downloading and extracting assests....", end="")

    # Downloading zip file using urllib package.
    urlretrieve(url, save_path)

    try:
        # Extracting zip file using the zipfile package.
        with ZipFile(save_path) as z:
            # Extract ZIP file contents in the same directory.
            z.extractall(os.path.split(save_path)[0])

        print("Done")

    except Exception as e:
        print("\nInvalid file.", e)
URL = r"https://www.dropbox.com/s/48hboi1m4crv1tl/opencv_bootcamp_assets_NB3.zip?dl=1"

asset_zip_path = os.path.join(os.getcwd(), f"opencv_bootcamp_assets_NB3.zip")

# Download if assest ZIP does not exists. 
if not os.path.exists(asset_zip_path):
    download_and_unzip(URL, asset_zip_path) 
```

```python
image=cv2.imread("Apollo_11_Launch.jpg",cv2.IMREAD_COLOR)

plt.imshow(image[:,:,::-1])
```

## Drawing a Line   
img=cv2.line(img,pt1,pt2,color,thickness,line_type,shift)   
This function has 4 required arguments:     
1-img:Image on which we will draw the line      
2-pt1:First point(x,y,location) of the line segment     
3-pt2:Second point of the line segment      
4-color:Color of the line which will be drawn   
Other optional arguments include:   
1-thickness:Integer specifying the line thickness. Default value is 1.  
2-lineType:Type of the line. 

```python
imageLine=image.copy()

#The line starts at (200,100) and ends at (400,100)
#Color of th eline is yellow
#Thickness of the line is 5px
#Line type is cv2.LINE_AA

cv2.line(imageLine,(200,100),(400,100),(0,255,255),thickness=5,lineType=cv2.LINE_AA)

plt.imshow(imageLine[:,:,::-1])
```

### Drawing a circle    
cv2.circle(img,center,radius,color,thickness,linetype,shift)    
1-img:Image on which we will draw a circle  
2-center:Center of the circle   
3-radius:radius of the circle   
4:color:Color of the circle which will be drawn

```python
imageCircle=image.copy()

cv2.circle(imageCircle,(600,550),100,(0,0,255),thickness=5,lineType=cv2.LINE_AA)

plt.imshow(imageCircle[:,:,::-1])
```

### Drawing a rectangle 
cv2.rectangle(img,pt1,pt2,color,thickness,lineType,shift)   
The 4 required arguments include:   
1-img:Image we want to draw the rectangle on    
2-pt1:Vertex of the rectangle. Usually we use the top-left vertex.  
3-pt2:Vertex of the rectangle opposite to pt1. Usually we use the bottom-right vertex here. 
4-color:Color of the rectangle  


```python
imageRectangle=image.copy()

cv2.rectangle(imageRectangle,(500,50),(700,650),(255,0,0),thickness=5,lineType=cv2.LINE_AA)

plt.imshow(imageRectangle[:,:,::-1])
```

### Adding text 
cv2.putText(img,text,org,fontface,fontScale,color,thickness,lineType,bottomLeftOrigin)  
The 6 required arguments include:   
1-img:Image on which the text is to be written  
2-text:Text string to be written    
3-org:Bottom-left corner of the text string in the image.   
4-fontface:Font type    
5-fontScale:Font scale factor that is multiplied by the font-specific base size     
6-color:font color

```python
imageText=image.copy()

text="Apollo 11 Saturn V Launch, July 16 1969"
fontScale=2.3
fontFace=cv2.FONT_HERSHEY_PLAIN
fontColor=(0,255,0)
fontThickness=2

cv2.putText(imageText,text,(200,30),fontFace,fontScale,fontColor,fontThickness,cv2.LINE_AA)

plt.imshow(imageText[:,:,::-1])
```
