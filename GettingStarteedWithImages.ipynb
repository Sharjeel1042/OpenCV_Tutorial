```python
import os #-o
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

from zipfile import ZipFile  #-o
from urllib.request import urlretrieve  #-o

from IPython.display import Image

%matplotlib inline
```

```python
#This whole cell block was used for downloading the images
def download_and_unzip(url, save_path):
    print(f"Downloading and extracting assests....", end="")

    # Downloading zip file using urllib package.
    urlretrieve(url, save_path)

    try:
        # Extracting zip file using the zipfile package.
        with ZipFile(save_path) as z:
            # Extract ZIP file contents in the same directory.
            z.extractall(os.path.split(save_path)[0])

        print("Done")

    except Exception as e:
        print("\nInvalid file.", e)
URL = r"https://www.dropbox.com/s/qhhlqcica1nvtaw/opencv_bootcamp_assets_NB1.zip?dl=1"

asset_zip_path = os.path.join(os.getcwd(), f"opencv_bootcamp_assets_NB1.zip")

# Download if assest ZIP does not exists. 
if not os.path.exists(asset_zip_path):
    download_and_unzip(URL, asset_zip_path)   
```

### Reading images using OpenCV

.imread(filename, flags)    
1-filename: Name of the image you want to read  
2-flags: In which format do you want to read the image, eg: greyscale, reg, etc. Default value is cv2.IMREAD_COLOR.  This is an optional argument

```python
#Reading image as greyscale
checkerBoard_Image=cv.imread('checkerboard_18x18.png',0)

#Printing the image data (pixel,value), element of a 2D numpy array
#Each pixel value is 8-bits[0,255]
print(checkerBoard_Image)
```

### Displaying image attributes

```python
#printing size of the image
print("Size of the small checkerboard image is ",checkerBoard_Image.shape)

#printing data-type of the image
print("Datatype of checkerboard image is: ",checkerBoard_Image.dtype)
```

### Displaying the image using matplotlib

```python
plt.imshow(checkerBoard_Image)
```

##### What happened?    
Even though the image was read as greyscale, it won't necessarily display in greyscale when using .imshow() as matplotlib uses different colormaps and it's possible that the grey scale color map is not set.

```python
#Setting colormap to greyscale for proper rendering
plt.imshow(checkerBoard_Image,cmap="gray")
```

### What happened?  
The color displayed above is different from the actucal image. This is because matplotlib expects the image to be in RGB fomrat whereas openCV reads the image in BGR format. Thus, for correct display, we need to reverse the channels of the image

```python
colaLogo_reversed=colaLogo_Img[:, :, ::-1]
plt.imshow(colaLogo_reversed)
```

# Splitting and merging color channels

cv2.split(): Divides a multi-channel array into several single-channel arrays   
cv2.merge(): Merges several arrays to make a single multi-channel array. All the input matrices must have the same size

```python
#Splitting the image into B,G,R components
img=cv.imread("New_Zealand_Lake.jpg",cv.IMREAD_COLOR)
b,g,r=cv.split(img)

#Show the channels
plt.figure(figsize=[20,5])
plt.subplot(141)
plt.imshow(b,cmap="gray")
plt.title("Blue Channel")


plt.figure(figsize=[20,5])
plt.subplot(142)
plt.imshow(g,cmap="gray")
plt.title("Green Channel")


plt.figure(figsize=[20,5])
plt.subplot(143)
plt.imshow(r,cmap="gray")
plt.title("Red Channel")


#Merging the images
imgMerged=cv.merge((b,g,r))
plt.subplot(144)
plt.imshow(imgMerged[:,:,::-1])
plt.title("Merged Output")
```

### Convertnig to different color spaces    
cv2.cvtColor(): Converts an image from one color space to another.

#### Function Syntax
dst=cv2.cvtColor(src,code)  
dst: Output image of same size and depth as src     
1-src: Input Image  
2-code: color space inversion code

### Changing from BGR to RGB

```python
imgNZ_Lake=cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.Figure(figsize=[20,5])
plt.subplot(141);plt.imshow(img);plt.title("Before Conversion")
plt.subplot(142);plt.imshow(imgNZ_Lake);plt.title("After Conversion")
```
